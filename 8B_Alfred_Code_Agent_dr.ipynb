In this zone, I engineer custom Python-based tools that extend the model beyond mere text generation and into functional task execution. Key Technical Implementations:

*   **Security & Isolation**: Restricting authorized imports (e.g., `datetime`, `time`) to ensure that agentic code execution remains performant and safe.

```python
# ZONE 3.1: The Skillset - Agent & Tool Initialization
from smolagents import CodeAgent, Tool, tool

# 1. TOOL DEFINITIONS: The "Instruction Manual" (Fixed Docstring)
@tool
def calculate_prep_time(guest_count: int, complexity: str) -> str:
    """
    Calculates preparation hours based on guest volume and dish complexity.

    Args:
        guest_count: The total number of guests invited to the mission event.
        complexity: The difficulty level of the catering ('standard' or 'high').
    """
    base_rate = 0.1 if complexity == "standard" else 0.2
    total_hours = guest_count * base_rate
    return f"Logistics Update: Total preparation required is {total_hours:.1f} hours."

class PartyThemeTool(Tool):
    name = "party_theme_generator"
    description = "Generates high-level creative concepts (noir, masquerade, futuristic)."
    inputs = {"category": {"type": "string", "description": "The vibe of the party."}}
    output_type = "string"

    def forward(self, category: str):
        themes = {"masquerade": "The Rogue's Gallery: A formal ball where masks hide true intentions."}
        return themes.get(category.lower(), f"A creative {category} concept.")

# 2. THE GOVERNOR: Infrastructure & Rate-Limit Logic
class TokenAwareGovernor:
    def __init__(self, tpm_limit=6000):
        self.tpm_limit = tpm_limit
        self.tokens_used_this_minute = 0
        self.last_reset_time = time.time()

    def estimate_tokens(self, prompt, guest_count):
        """FORENSIC ESTIMATION: Calculates total TPM impact before execution."""
        base_load = 750
        prompt_tokens = int(len(prompt.split()) * 1.35)
        complexity_buffer = (guest_count // 50) * 50
        expected_output = 350 + complexity_buffer
        total_estimate = base_load + prompt_tokens + expected_output
        print(f"ğŸ“Š Forensic Audit: Base({base_load}) + Prompt({prompt_tokens}) + Output({expected_output})")
        return total_estimate

    def pre_flight_clearance(self, prompt, guest_count):
        current_time = time.time()
        if current_time - self.last_reset_time > 60:
            self.tokens_used_this_minute = 0
            self.last_reset_time = current_time

        estimated = self.estimate_tokens(prompt, guest_count)
        if (self.tokens_used_this_minute + estimated) > self.tpm_limit:
            wait_needed = 60 - (current_time - self.last_reset_time) + 2
            print(f"ğŸ›‘ PROACTIVE GOVERNANCE: Party for {guest_count} guests exceeds TPM budget.")
            print(f"â³ Alfred is pausing for {wait_needed:.1f}s to reset LPU bucket...")
            time.sleep(max(0, wait_needed))
            self.tokens_used_this_minute = 0
            self.last_reset_time = time.time()
        return estimated

    def parse_429(self, error_msg):
        print(f"Governor Intercepted 429: {error_msg}")
        match = re.search(r"try again in (?:(\d+)h)?(?:(\d+)m)?([\d\.]+)s", error_msg)
        if match:
            h = float(match.group(1)) if match.group(1) else 0
            m = float(match.group(2)) if match.group(2) else 0
            s = float(match.group(3)) if match.group(3) else 0
            return (h * 3600) + (m * 60) + s + 2
        return 60

# 3. INITIALIZATION: Skillset and Governance Online
governor = TokenAwareGovernor(tpm_limit=6000)

agent = CodeAgent(
    tools=[calculate_prep_time, PartyThemeTool()],
    model=model,
    additional_authorized_imports=['datetime', 'time'],
    max_steps=10
)

print("âœ… Zone 3.1 Complete: Alfred's Skillset (Tools) is initialized and tools are registered.")
```

## ZONE 3.2: The Governor - Infrastructure & Rate-Limit Governance (Control)
**Objective**: To implement a deterministic safety layer that sits between the Agentâ€™s reasoning and the hardwareâ€™s physical constraints. In this zone, I developed a custom `TokenAwareGovernor` designed to *hear* and interpret infrastructure signalsâ€”specifically the *429 Rate Limit* errors common in high-velocity LPU environments.

**Why this is critical**: Agentic workflows are inherently non-deterministic; a single creative loop can exponentially bloat the token count, leading to system-wide *choking* or service suspension. My Governor prevents this through:

*   **Forensic Pre-Flight Estimation**: Calculating the total TPM (Tokens Per Minute) impact before a request is fired, ensuring I never cross the 6,000-token threshold.

*   **Sliding-Window Memory**: Maintaining a real-time record of usage to enforce proactive pauses, rather than relying on reactive error handling.

*   **Automated Backoff Parsing**: Utilizing regex-based forensic parsing to interpret exact cooldown requirements from API headers, turning *failure* into a managed temporal pause.

```python
## ZONE 3.2: The Governor - Infrastructure & Rate-Limit Governance

import re
import time

class TokenAwareGovernor:
    def __init__(self, tpm_limit=6000):
        self.tpm_limit = tpm_limit
        self.tokens_used_this_minute = 0
        self.last_reset_time = time.time()

    def estimate_tokens(self, prompt, guest_count):
        """
        FORENSIC ESTIMATION: Calculates total TPM impact before execution.
        """
        # 1. FIXED BASE LOAD (smolagents Tool Definitions)
        base_load = 750

        # 2. VARIABLE PROMPT WEIGHT
        prompt_tokens = int(len(prompt.split()) * 1.35)

        # 3. ANTICIPATED OUTPUT (Scaling with GUEST_COUNT)
        complexity_buffer = (guest_count // 50) * 50
        expected_output = 350 + complexity_buffer

        total_estimate = base_load + prompt_tokens + expected_output

        print(f"ğŸ“Š Forensic Audit: Base({base_load}) + Prompt({prompt_tokens}) + Output({expected_output})")
        return total_estimate

    def pre_flight_clearance(self, prompt, guest_count):
        """Checks the LPU bucket before sending the request."""
        current_time = time.time()

        # 1. SLIDING WINDOW RESET
        if current_time - self.last_reset_time > 60:
            self.tokens_used_this_minute = 0
            self.last_reset_time = current_time

        # 2. FORENSIC ESTIMATION
        estimated = self.estimate_tokens(prompt, guest_count)

        # 3. PROACTIVE INTERCEPT
        if (self.tokens_used_this_minute + estimated) > self.tpm_limit:
            wait_needed = 60 - (current_time - self.last_reset_time) + 2
            print(f"ğŸ›‘ PROACTIVE GOVERNANCE: Party for {guest_count} guests exceeds TPM budget.")
            print(f"â³ Alfred is pausing for {wait_needed:.1f}s to reset LPU bucket...")
            time.sleep(max(0, wait_needed))

            # Reset after the wait
            self.tokens_used_this_minute = 0
            self.last_reset_time = time.time()

        return estimated

    def parse_429(self, error_msg):
        """Upgraded regex logic for daily/hourly limits."""
        print(f"Governor Intercepted 429: {error_msg}")
        match = re.search(r"try again in (?:(\d+)h)?(?:(\d+)m)?([\d\.]+)s", error_msg)
        if match:
            h = float(match.group(1)) if match.group(1) else 0
            m = float(match.group(2)) if match.group(2) else 0
            s = float(match.group(3)) if match.group(3) else 0
            total_wait = (h * 3600) + (m * 60) + s + 2
            return total_wait
        return 60

# Initialize for the 8B Model
governor = TokenAwareGovernor(tpm_limit=6000)

# TEST THE LOGIC (Class Version)
test_error = 'Rate limit reached... Please try again in 22m37.34s.'
wait_time = governor.parse_429(test_error)
print(f"Test Result: Governor recommends sleep for {wait_time} seconds")
```

##ZONE 5: THE FORENSIC DASHBOARD & HEALING AUDIT

In this layer, I visualize the mission's performance. I calculate the Healing Cost Multiplier and Forensic Variance, providing a clear picture of how much *extra* work the system performed to reach a successful state.

1. The Estimate is formulated in the `TokenAwareGovernor` class of my Zone 3.2.  It uses a *Pre-Flight* formula to guess the token cost before Alfred even speaks. Here is the math I used for the 8B run:

   *   **Base Load**: 750 (Fixed cost for tool definitions and Alfred's persona).


   *   **Prompt Weight**: 148 (The word count of my mission prompt $\times 1.35$).
   *   **Expected Output**: 500 (A buffer for the playlist and menu).
   *   **Total Estimate**: 1,398 tokens.

   **Detailed explanation for above labels**:
   
   **Base Load**: `smolagents` System Prompt + Tool Schemas.
   
      *   **The Logic**: When I initialize a `CodeAgent`, the framework automatically prepends a massive "System Message" that explains how to use Python as a tool.

      **Prompt Weight**: Standard Token-to-Word Ratios (1.35x multiplier).
      *   **The Logic**: LLMs don't read words; they read tokens. For technical prompts with code snippets or specific instructions, the ratio is typically higher than standard English (which is roughly 1.3x).
      
      *   **The Calculation**: I took my mission prompt (approx. 110 words) and applied a 1.35x safety multiplier to account for punctuation and formatting tokens; so, $110 \times 1.35 \approx 148$.
      
      **The Expected Output**: *Worst-Case* Mission Output Density.
      *   **The Logic**: I measured the character count of a standard 3-course menu and a 10-song playlist.

      *   **The Calculation**: A high-quality response for a 150-guest mission typically generates between 400 and 600 tokens. I chose 500 as the *Safety Ceiling* for the governor. If the model goes over this, my Lexical De-confliction guardrails are supposed to kick in.

2. **The Reality**: The *Actual* Count

   **Why the massive jump?** Because the 8B model didn't just give a ***one-shot*** answer. It entered a ***ReAct loop** (Step 1, Step 2, Step 3)*. Every time the model *thinks* or *calls a tool*, it repeats the previous steps in its memory. My Governor estimated a **Single-Turn interaction**, but the agent performed a **Multi-Step Mission**.
   
  3. **The Formulas: Accuracy vs. Variance**
  
      To calculate these for the dashboard, I used standard forensic accounting logic:
      
      **The Forensic Variance (+87.9%)**
      
      This represents how much the model overran the budget.
      <br>
      $$\text{Variance %} = \left( \frac{\text{Actual} - \text{Estimate}}{\text{Actual}} \right) \times 100$$$${11,542 - 1,398}/{11,542} x 100 ={87.9\%}$$
