## Zone 1: Orchestration & Observability Vault

**Objective**: Establish a secure *source of truth* for credentials and initialize the OpenTelemetry (OTEL) bridge.

In this section, I implement three critical enterprise-level pillars:

1.   **Secret Management**: Utilize the Google Colab `userdata` utility. This ensures that sensitive API keys for `Groq` and `Langfuse` are never hard-coded or exported to the notebook's `GitHub`/Hub history.
2.   **Telemetry Instrumentation**: Activate the `SmolagentsInstrumentor`. This *wraps* the agent's internal processes, allowing us to capture hidden reasoning steps, tool latency, and exact prompt/completion pairs.
3.   **`gRPC` Bridge**: Configure the OpenTelemetry exporter to stream data to Langfuse via gRPC. By using Base64-encoded Basic Auth in the headers, I maintain a secure, high-speed persistent connection to the US Cloud.

Also, in pursuit of Production-Grade Observability, I have refined the gRPC handshake as follows.

**Header Normalization**: I've switched to a standard Basic {token} format for the OTEL_EXPORTER_OTLP_HEADERS. This ensures maximum compatibility with the OpenTelemetry collector.

**Deterministic Endpoints**: I have strictly defined the OTLP endpoint to ensure the binary Protobuf data reaches the correct ingestion gateway without redirection.

```python
# ZONE 1: The Vault - Authentication & Verified Observability Setup
import os, base64
from google.colab import userdata
from langfuse import get_client
from smolagents import LiteLLMModel
from openinference.instrumentation.smolagents import SmolagentsInstrumentor

# 1. CREDENTIAL INJECTION
os.environ["GROQ_API_KEY"] = userdata.get("GROQ_API_KEY")
os.environ["LANGFUSE_PUBLIC_KEY"] = userdata.get("LANGFUSE_PUBLIC_KEY")
os.environ["LANGFUSE_SECRET_KEY"] = userdata.get("LANGFUSE_SECRET_KEY")
os.environ["LANGFUSE_HOST"] = "https://us.cloud.langfuse.com"

# 2. MODEL GOVERNOR (The 429 Fix)
# I initialize the model with strict pacing to protect my Groq TPM/RPM quota.
model = LiteLLMModel(
    model_id="groq/llama-3.3-70b-versatile",
    api_key=os.environ["GROQ_API_KEY"],
    requests_per_minute=20,  # Pacing Alfred to avoid the 12k TPM limit
    retry=True               # Enables smart backoff if limits are hit
)

# 3. LANGFUSE CLIENT INITIALIZATION & PRE-FLIGHT AUTH CHECK
langfuse = get_client()

if langfuse.auth_check():
    print("âœ… Vault Verified: Langfuse Authentication Successful.")
else:
    print("âŒ Vault Error: Authentication failed. Please check your Secret Keys.")

# 4. OPENTELEMETRY (OTEL) CONFIGURATION
auth_str = f"{os.environ['LANGFUSE_PUBLIC_KEY']}:{os.environ['LANGFUSE_SECRET_KEY']}"
encoded_auth = base64.b64encode(auth_str.encode()).decode()

os.environ["OTEL_EXPORTER_OTLP_ENDPOINT"] = f"{os.environ['LANGFUSE_HOST']}/api/public/otel"
os.environ["OTEL_EXPORTER_OTLP_HEADERS"] = f"Authorization=Basic {encoded_auth}"

# 5. START THE OBSERVER
try:
    SmolagentsInstrumentor().uninstrument()
except Exception:
    pass

SmolagentsInstrumentor().instrument()

print("âœ… Zone 1 Complete: The Engine is governed and monitored by Langfuse via gRPC.")
```

The warning `Attempting to uninstrument while already uninstrumented` is a byproduct of an **Idempotent Reset Pattern**. By calling `.uninstrument()` before `.instrument()`, I ensure a *clean slate* for the telemetry bridge. This prevents duplicate instrumentation layers (which could cause double-counting of tokens or metrics) if the cell is re-run during development. The warning simply confirms the system was already clean, which is the desired starting state for a reliable monitoring loop.

##Zone 3: The Skillset (Extensible Tool Architecture)
**Objective**: To equip the agent with specialized tools for logistics, creativity, and self-regulation.

In this section, I transition from the "Brain" to the "Hands." I implement the Strategy Pattern by providing Alfred, the agent, with a suite of tools it can use to solve the mission. Key highlights include:

**Logical Computation**: The `calculate_prep_time` tool handles deterministic math, ensuring that Alfred doesn't "hallucinate" timing for the gala.

**Creative Synthesis**: Instead of hard-coding a menu or playlist, I provide a `party_theme_generator` that allows Alfred to synthesize unique ideas based on its internal knowledge base.

**Self-Regulation (The "Cool-Down" Pattern)**: I include the alfred_pause tool. This is a sophisticated engineering touch that allows the agent to programmatically waitâ€”-therefore, helping Alfred stay within the Resource Governor limits during high-complexity missions.

```python
# ZONE 3: The Skillset - Custom Tools & Agent Initialization
from smolagents import CodeAgent, Tool, tool

@tool
def calculate_prep_time(guest_count: int, complexity: str) -> str:
    """
    Calculates the required preparation time based on guest volume and event depth.
    Args:
        guest_count: The number of expected attendees.
        complexity: 'high' for multi-course/themed, 'standard' for casual.
    """
    # Logic: 0.5 hours per 10 guests for standard, 1 hour per 10 for high complexity
    base_rate = 0.1 if complexity == "standard" else 0.2
    total_hours = guest_count * base_rate
    return f"Logistics Update: Total preparation required is {total_hours:.1f} hours for {guest_count} guests."

@tool
def alfred_pause(seconds: int) -> str:
    """
    A self-regulation tool that allows the agent to pause execution.
    Args:
        seconds: Number of seconds to wait before the next reasoning step.
    """
    import time
    time.sleep(seconds)
    return f"System: Alfred has paused for {seconds}s to respect the Resource Governor."

class PartyThemeTool(Tool):
    name = "party_theme_generator"
    description = "Generates high-level creative concepts for mansion-scale events."
    inputs = {
        "category": {
            "type": "string",
            "description": "The vibe of the party (e.g., 'noir', 'masquerade', 'futuristic').",
        }
    }
    output_type = "string"

    def forward(self, category: str):
        # I allow the LLM to 'think' through the themes, but provide a structured fallback
        themes = {
            "noir": "Shadows of Gotham: A 1940s jazz-infused mystery evening.",
            "masquerade": "The Rogue's Gallery: A formal ball where masks hide true intentions.",
            "futuristic": "Neo-Gotham Gala: High-tech, neon-lit celebration of the future."
        }
        return themes.get(category.lower(), f"A creative {category} concept generated by the Wayne database.")

# INITIALIZE THE AGENT
# I include 'time' and 'datetime' in authorized imports for the agent's internal code blocks.
agent = CodeAgent(
    tools=[calculate_prep_time, alfred_pause, PartyThemeTool()],
    model=model,
    additional_authorized_imports=['datetime', 'time', 'numpy'],
    max_steps=10,
    verbosity_level=2
)

print("âœ… Zone 3 Complete: Alfred's Skillset is initialized and tools are registered.")
```

##Diagnostic: Alfred's Forensic Hearing Test
**Objective**: To verify the robustness of my Regular Expression (re) parser before executing the final mission.

The diagnostic test uses **Equivalence Partitioning**. By testing a 'seconds-only' string, a 'minutes-seconds' string, and an 'hours-minutes-seconds' string, I ensure the mathematical conversion is robust regardless of the severity of the rate limit. Because the **Groq API** returns rate-limit errors in a variety of string formats (seconds only, minutes and seconds, or hours), I must ensure my backoff logic can surgically extract wait times from raw text. This *sandbox* simulates three common bottleneck scenarios to confirm Alfred can correctly convert API feedback into actionable wait-timers. It ensures that the `time.sleep()` logic receives a valid number, rather than a string, preventing runtime crashes.

*   **Scenario A**: Standard Minute-based limit (TPM)
*   **Scenario B**: Daily rolling window reset (TPD)
*  **Scenario C**: Extreme edge-case (Hours)

**Technical Verification**
The code below uses the following pattern to capture numeric groups: `r"(?:(\d+)h)?(?:(\d+)m)?([\d\.]+)s"`

*  Non-capturing groups (`?:`): I check for hours and minutes without making them mandatory
*  Digit capture (`\d+`): I isolate the raw numbers for mathematical conversion
*   Float support (`[\d\.]+`): I capture the precise milliseconds provided by the API for high-accuracy pauses







```python
import re
import time

# Diagnostic Hearing Test
def forensic_governor(error_message):
    """
    Parses the Groq 429 error message to extract the wait time.
    Handles formats like '22m37.343s' or just '5.28s'.
    """
    print(f"Governor Intercepted Error: {error_message}")

    # Regex to find the 'Please try again in ...' pattern
    # This captures minutes and seconds, or just seconds
    match = re.search(r"try again in (?:(\d+)m)?([\d.]+)s", error_message)

    if match:
        minutes = float(match.group(1)) if match.group(1) else 0
        seconds = float(match.group(2))
        total_wait = (minutes * 60) + seconds

        print(f"âœ… Governor parsed wait time: {total_wait} seconds.")
        return total_wait
    else:
        print("âš ï¸ Governor could not parse exact time. Defaulting to 60s backoff.")
        return 60

# TEST THE LOGIC
test_error = 'Rate limit reached... Please try again in 22m37.343999999s.'
wait_time = forensic_governor(test_error)
print(f"Test Result: Sleep for {wait_time} seconds")
```

```python
# ZONE 4: The Mission - Final Execution with Public Portfolio Trace
from langfuse import propagate_attributes
import time
import re

# 1. MISSION METADATA & HARDCODED PORTFOLIO LINKS
# I have pivoted to the 8B model to demonstrate high-speed, cost-effective orchestration.
SESSION_ID = "alfred-production-run-v13"
USER_ID = "portfolio-viewer"
GUEST_COUNT = 150
TPM_LIMIT_V = 12000   # Tokens Per Minute (Hardware Constraint). Use for versatile model.
TPD_LIMIT_V = 100000  # Tokens Per Day (Daily Quota). Use for versatile model.
TPM_LIMIT_I = 6000   # Tokens Per Minute (Hardware Constraint). Use for instant model.
TPD_LIMIT_I = 500000  # Tokens Per Day (Daily Quota). Use for instant model.

# --- PUBLIC LINK CONFIGURATION ---
LANGFUSE_PROJECT_ID = "cmjc4u4da009iad07f0mh1f8n"
LANGFUSE_BASE_URL = "https://us.cloud.langfuse.com"

# Pricing inputs for the selected model
MODEL_PRICING = {
    "groq/llama-3.1-8b-instant": {"input": 0.05, "output": 0.08},
    "default": {"input": 0.05, "output": 0.08}
}

# 2. RESET MONITOR & EXECUTE MISSION
# I am ensuring the agent is using the 8B engine for this specific run.
agent.model.model_id = "llama-3.1-8b-instant"

if hasattr(agent, 'monitor'):
    agent.monitor.reset()

# Setting variables to 'None' creates a "placeholder" in computer's memory, whereby computer knows
# these variables exist, even if they don't have a value yet--it's a safety net to prevent crashing.
captured_trace_id = None
final_report = None

try:
    # propagate_attributes stitches Alfred's tool calls into a unified narrative.
    with propagate_attributes(session_id=SESSION_ID, user_id=USER_ID):
        with langfuse.start_as_current_observation(name="Alfred_Upbeat_Gala") as observation:
            # Alfred will now process the multi-decade musical requirement.
            captured_trace_id = langfuse.get_current_trace_id() # Capturing the ID so I can generate the public URL later.

            mission_prompt = f"""
            Alfred, finalize the 'Villain Masquerade' plan at Wayne Manor for {GUEST_COUNT} guests.
            Your FINAL ANSWER must include:
            1. Prep time for {GUEST_COUNT} guests.
            2. The Theme (one sentence).
            3. The full 5-song Playlist (Artist - Title) with upbeat sound and can span from the 1950s to the present day.
            Choose one definitive high-energy hit for each major era across the time span to keep guests dancing.
            4. The 3-course Menu with MICRO-DESCRIPTIONS.
               Example: 'Poison Apple: Tart granny smith glazed in dark caramel.'
            """

            final_report = agent.run(mission_prompt)

            # Makes this specific trace public for my portfolio viewers.
            observation.update(output=str(final_report), public=True)

except Exception as e:
    error_msg = str(e)
    # DEFENSIVE ENGINEERING: THE GOVERNOR
    # Even on the 8B model, I keep my Governor active to handle unexpected provider jitter.
    if "rate_limit_exceeded" in error_msg or "429" in error_msg:
        match = re.search(r"try again in (?:(\d+)h)?(?:(\d+)m)?([\d\.]+)s", error_msg)
        if match:
            h = float(match.group(1)) if match.group(1) else 0
            m = float(match.group(2)) if match.group(2) else 0
            s = float(match.group(3)) if match.group(3) else 0

            wait_time = (h * 3600) + (m * 60) + s + 2
            print(f"â³ Rate Limit Detected. Alfred is pausing for {wait_time:.1f}s...")
            time.sleep(wait_time)
    else:
        print(f"âŒ Unexpected Error: {e}")

# 3. DATA SYNC (The Sync Gate)
langfuse.flush()
time.sleep(1)

# 4. THE EFFICIENCY AUDIT
input_tokens = getattr(agent.monitor, 'total_input_token_count', 0)
output_tokens = getattr(agent.monitor, 'total_output_token_count', 0)
total_tokens = input_tokens + output_tokens

# Identifying the model ID from the agent's actual state
current_model_id = getattr(agent.model, 'model_id', 'default')
rates = MODEL_PRICING.get(current_model_id, MODEL_PRICING["default"])

cost_in = (input_tokens / 1e6) * rates["input"]
cost_out = (output_tokens / 1e6) * rates["output"]
total_cost = cost_in + cost_out

# Efficiency Visualization Logic
tpm_ratio = total_tokens / TPM_LIMIT_I
tpd_ratio = total_tokens / TPD_LIMIT_I
status_msg = "âš ï¸ QUOTA EXHAUSTED" if (tpd_ratio > 0.98) else "âœ… STABLE"

def get_bar(ratio):
    length = 20
    filled = int(length * min(ratio, 1.0))
    return "â–ˆ" * filled + "-" * (length - filled)

# 5. CONSOLIDATED ANALYTICS DASHBOARD
print("\n" + "="*55)
print(f"ğŸ›¡ï¸  MISSION ANALYTICS | {status_msg}")
print(f"User ID: {USER_ID} | Model: {current_model_id}")
print(f"TPM Usage: [{get_bar(tpm_ratio)}] {tpm_ratio*100:.1f}%")
print(f"TPD Usage: [{get_bar(tpd_ratio)}] {tpd_ratio*100:.1f}%")
print("-" * 55)
print(f"Total Tokens: {total_tokens:,}")
print(f"Input Cost:   ${cost_in:.6f}")
print(f"Output Cost:  ${cost_out:.6f}")
print(f"Run Total:    ${total_cost:.5f}")
print(f"Guest Cost:   ${total_cost/GUEST_COUNT:.6f} / guest")
print("-" * 55)

if captured_trace_id:
    trace_url = f"{LANGFUSE_BASE_URL}/project/{LANGFUSE_PROJECT_ID}/traces/{captured_trace_id}"
    print(f"Verified Trace: {trace_url}")
else:
    print("Verified Trace: Telemetry Sync Pending")
print("="*55)

if final_report:
    print(f"\n--- ALFRED'S FINAL REPORT ---\n\n{final_report}")
```
